```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('D:/RAKAN/Data Python/Group3_Project/Data/clean_fifa_players.csv')
dataset
```

```python
df = dataset
```

```python

df.drop('height_cm', axis=1, inplace=True)
df.drop('weight_kgs', axis=1, inplace=True)
df.drop('weak_foot(1-5)', axis=1, inplace=True)
df.drop('skill_moves(1-5)', axis=1, inplace=True)
df.drop('body_type', axis=1, inplace=True)
```

```python
df
```

```python
X = df.iloc[:, 4:]
X
```

```python
Y = df['positions']
Y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm_scaler = MinMaxScaler()
```

```python
X = mm_scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .1)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
```

```python
Gaus = GaussianNB()
Bernou = BernoulliNB()
Multinom = MultinomialNB()
```

```python
Gaus.fit(X_train, Y_train)
Gaus.score(X_test, Y_test)
```

```python
Bernou.fit(X_train, Y_train)
Bernou.score(X_test, Y_test)
```

```python
Multinom.fit(X_train, Y_train)
Multinom.score(X_test, Y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
NN = KNeighborsClassifier(n_neighbors=5)
svc = SVC(probability=True)
```

```python
NN.fit(X_train, Y_train)
```

```python
NN.score(X_test, Y_test)
```

```python
svc.fit(X_train, Y_train)
```

```python
a = svc.predict_proba([X_test[6]])
a
```

```python
svc.predict([X_test[5]])
```
